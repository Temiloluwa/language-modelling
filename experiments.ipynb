```python
import os
import nltk
from utils import read_text_file
import tensorflow as tf
import re
```

```python
data_path = os.path.join(os.getcwd(), 'datasets', 'shakespeare.txt')
list_of_lines = read_text_file(data_path)
list_of_lines = [re.sub(r"[:;,.?!]", lambda x: " "+x.group(0), i) for i in list_of_lines]
tokenizer = tf.keras.preprocessing.text.Tokenizer(filters='"#$%&()*+-/<=>@[\\]^_`{|}~\t\n')
tokenizer.fit_on_texts(list_of_lines)
tokenized_lines = tokenizer.texts_to_sequences(list_of_lines)
index_to_word_mapping = {v:k for k,v in tokenizer.word_index.items()}
```
